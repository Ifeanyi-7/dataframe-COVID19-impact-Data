```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
# Creating a more detailed DataFrame with COVID-19 Impact Data
```

```python
data = {
    "Region": ["Global", "USA", "Europe", "Asia", "Africa", "South America"],
    "Total Cases (Millions)": [600, 98, 150, 250, 20, 82],
    "Total Deaths (Millions)": [6.5, 1.1, 2.3, 2.5, 0.3, 1.2],
    "Economic Loss (Trillions $)": [12, 4.5, 3.2, 2.8, 0.5, 1],
    "Jobs Lost (Millions)": [114, 30, 35, 28, 10, 11],
    "Education Disruptions (Months)": [18, 12, 14, 20, 24, 16],
    "Telemedicine Growth (%)": [350, 500, 400, 300, 250, 270],
    "Remote Work Adoption (%)": [85, 90, 80, 75, 60, 70],
    "E-commerce Growth (%)": [45, 50, 42, 48, 30, 38]
}


```

```python
df = pd.DataFrame(data)
```

```python
print(data)
```

```python
print(df.head())
```

```python
# Data Summary
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Handling Missing Values (if any)
```

```python
df.fillna(0, inplace=True)
```

```python
# Visualization 1: Bar Plot
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(y="Region", x="Total Cases (Millions)", data=df, palette="magma")
plt.xlabel("Total Cases (Millions)")
plt.ylabel("Region")
plt.title("COVID-19 Cases by Region")
plt.show()
```

```python
# Visualization 2: Pie Chart - Economic Loss
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df["Economic Loss (Trillions $)"], labels=df["Region"], autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title("Economic Loss Distribution by Region")
plt.show()
```

```python
# Visualization 3: Line Plot - Telemedicine Growth
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=df["Region"], y=df["Telemedicine Growth (%)"], marker="o", color="green")
plt.xlabel("Region")
plt.ylabel("Telemedicine Growth (%)")
plt.title("Telemedicine Growth During COVID-19")
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Matrix
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.iloc[:, 1:].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of COVID-19 Impact")
plt.show()

```
